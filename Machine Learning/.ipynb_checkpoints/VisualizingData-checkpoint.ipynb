```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("housing.csv")
```

```python
df['income_cat'] = pd.cut(df['median_income'], bins = [0, 1.5, 3.0, 4.5, 6.0, np.inf], labels = [1,2,3,4,5])
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)
for train_index, test_index in split.split(df, df['income_cat']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
# lets remove the income_cat column
for sett in (strat_test_set, strat_train_set):
    sett.drop('income_cat', axis = 1, inplace = True)
```

```python
df = strat_train_set.copy()
```

```python
df.plot(kind = 'scatter', x = 'latitude', y = 'longitude', grid = True, cmap = 'jet', c = 'median_house_value')
```

```python
df.drop(labels = 'ocean_proximity', inplace = True, axis = 1)
df.corr()
```

```python
from pandas.plotting import scatter_matrix
attributes = ['median_income', 'housing_median_age', 'median_house_value']
scatter_matrix(df[attributes], figsize = (12,9))
```

### These plots help us understand how two attributes are related to each other

#### Thus, we can infer that 'median_income' repesents the 'median_house_value'(label) better than other attributes
