```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
# a = requests.get("https://books.toscrape.com/")
# print(a.text)
```

```python
for i in range(1,51):
    a = requests.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
    with open(f"htmls/page{i}.html","w",encoding="utf-8") as f:
        f.write(a.text)  
    print(f"Downloaded Page {i}")
```

```python
items = []
for i in range(1,51):
    with open(f"htmls/page{i}.html", encoding="utf-8") as f:
            content = f.read()
    soup = bs(content, "html.parser")
    articles = soup.select("article.product_pod")   #selects all the articles 
    for article in articles:
        title = article.find("h3").find("a")["title"]     # articles > h3 > a > title
        price = article.select_one("p.price_color").text.split("£")[1]
        stock = article.select_one("p.instock").text.strip()
        rating_ele = article.select_one("p.star-rating")
        rating = rating_ele["class"][1]
        # print(f"Title: {title} \t Price: {price} \t Rating: {rating}")  
        items.append([title,price,rating,stock])
# items    prints all the items
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(items, columns = ["Book","Price","Rating","Stock"])
df
```

```python
df.to_csv("Books_details.csv",index = False)
```
